# 1. Deskripsi data

Data yang digunakan adalah data Tingkat Pengangguran Terbuka (TPT) disetiap kabupaten lampung pada tahun 2022 dan 2023.

Sumber data: Badan Pusat Statistik Provinsi Lampung

# 2. Tujuan

Tujuan dari visualisasi ini adalah untuk mengetahui perbandingan tingkat pengangguran terbuka pada tahun 2022 dan 2023 di setiap kabupaten Provinsi lampung.

#3. User
A boardroom environment: data ini bisa digunakan oleh sekelompok pejabat senior atau pembuat kebijakan yang memiliki pengetahuan mendalam mengenai kondisi ekonomi dan ketenagakerjaan regional.

#4. Fungsi
Visualisasi ini memiliki fungsi komparatif, yakni untuk membandingkan tingkat pengangguran di kabupaten-kabupaten Provinsi Lampung antara tahun 2022 dan 2023. Ini memudahkan user untuk melihat apakah pengangguran meningkat atau menurun di setiap wilayah. Visualisasi ini dapat dijadikan sebagai alat bantu dalam pembuatan kebijakan dalam membuat keputusan strategis terkait penanggulangan pengangguran di daerah. Visualisasi ini juga akan membantu dalam menentukan wilayah-wilayah yang perlu diprioritaskan untuk intervensi kebijakan, serta mengevaluasi kebijakan yang berhasil atau perlu diperbaiki.


Berikut kode dan visualisasi dari data yang digunakan

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
file_path = 'Tingkat Pengangguran Terbuka (TPT), 2023.xlsx'
data = pd.read_excel(file_path)

# Rename columns for clarity
data_clean = data.rename(columns={'Wilayah': 'Kabupaten', 'Unnamed: 1': 'TPT_2023', 'Unnamed: 2': 'TPT_2022'})

# Remove rows where Kabupaten is NaN (the first two rows seem to be headers or irrelevant)
data_clean = data_clean.dropna(subset=['Kabupaten'])

# Convert TPT_2023 and TPT_2022 to numeric
data_clean['TPT_2023'] = pd.to_numeric(data_clean['TPT_2023'], errors='coerce')
data_clean['TPT_2022'] = pd.to_numeric(data_clean['TPT_2022'], errors='coerce')

data.head()
```

```python
# Data for plotting
kabupaten = data_clean['Kabupaten']
tpt_2023 = data_clean['TPT_2023']
tpt_2022 = data_clean['TPT_2022']

# Setting up the bar positions
bar_width = 0.35
index = np.arange(len(kabupaten))

# Plotting
plt.figure(figsize=(12, 6))
bar1 = plt.bar(index, tpt_2023, bar_width, label='2023', color='b')
bar2 = plt.bar(index + bar_width, tpt_2022, bar_width, label='2022', color='g')

# Labels and titles
plt.xlabel('Kabupaten')
plt.ylabel('Tingkat Pengangguran (%)')
plt.title('Perbandingan Tingkat Pengangguran Terbuka di Kabupaten Provinsi Lampung (2022 vs 2023)')
plt.xticks(index + bar_width / 2, kabupaten, rotation=90)
plt.legend()

# Layout and saving
plt.tight_layout()
plt.savefig('perbandingan_pengangguran_2022_vs_2023.png')
plt.show()
```

>

Secara keseluruhan, dari visualisasi di atas, Provinsi Lampung menunjukkan tingkat pengangguran di beberapa kabupaten menurun dari tahun sebelumnya. Namun, meskipun ada beberapa perbaikan, beberapa daerah seperti Pesisir Barat tetap memiliki tingkat pengangguran yang tinggi pada tahun 2023. Ini menunjukkan bahwa beberapa wilayah masih menghadapi tantangan dalam penyerapan tenaga kerja. Kabupaten Bandar Lampung dan Metro juga mencatat tingkat pengangguran yang relatif tinggi dibandingkan daerah lain di Provinsi Lampung. Hal tersebut memerlukan perhatian lebih untuk mengatasi masalah pengangguran. Analisis ini dapat menjadi acuan bagi pemerintah daerah dalam merumuskan kebijakan yang lebih tepat untuk mengurangi pengangguran di daerah-daerah yang memerlukan intervensi lebih lanjut.
